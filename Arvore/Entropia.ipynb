# Entropia

A entropia é uma grandeza que mede a desordem, tanto de objetos físicos quanto de informações. Quanto maior o grau da entropia maior é a desordem e, quanto menor o grau da entropia melhor a organização.

## Medindo a entropia de uma base de dados

A equação abaixo exibe a fórmula de cálculo da entropia para uma base de dados

```python
from IPython.display import Math
Math(r'Entropia(Y) = {-\sum \limits _{i=1} ^{N} p(y_{i})*log2(p(y_{i}))}')
```

Onde

    - Y é a variável alvo
    - N é o total de classes (valores que variável pode assumir
    - p(y) é a probabilidade a priori da classe

# Exemplo

```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('dif_dist_dataset.csv')
```

```python
dados
```

```python
dados['Perc_Classe_1'] = dados.Classe_1 / 100
dados['Perc_Classe_2'] = dados.Classe_2 / 100
dados['Entropia']      = -1 * (dados.Perc_Classe_1*np.log2(dados.Perc_Classe_1) + dados.Perc_Classe_2*np.log2(dados.Perc_Classe_2))
```

```python
dados
```

## Análise Gráfica da Entropia

```python
from matplotlib import pyplot as plty
```

```python
plty.plot(dados.Perc_Classe_1, dados.Entropia)
plty.xlabel('% da classe 1 (% da classe 2 é o complemento)')
plty.ylabel('Entropia')
plty.show()
```
